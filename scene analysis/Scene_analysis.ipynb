```python
import io
import pandas as pd
import requests  
from bs4 import BeautifulSoup
primary_url = 'https://fangj.github.io/friends/'
request=requests.get(primary_url)
soup=BeautifulSoup(request.content)
urllist=[]
for i in soup.find_all('a'):
    urllist.append(i.get('href'))
urllist=[primary_url+url for url in urllist]
```

```python
import re
scene_list=[]
text_list=[]
def processing_urls(urllist):
    for url in urllist:        
        newreq=requests.get(url)
        
        isoup=BeautifulSoup(newreq.content)
        text_list.append(isoup.find_all('p'))
        
processing_urls(urllist)
def scenes(text_list):
    for i in text_list:
        for j in i:
            if(re.search(r'Scene',str(j))):
                scene_list.append(str(j))
scenes(text_list)                
```

```python
clean_scenes_list = []

for i in scene_list:
    pattern = re.search(r'\[(.*?)\]',i)
    if pattern:
        pattern = pattern.group()[1:-1]
        clean_scenes_list.append(pattern) 
clean_scenes_list        
```
