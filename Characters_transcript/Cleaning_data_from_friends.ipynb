```python
import requests  
from bs4 import BeautifulSoup
primary_url = 'https://fangj.github.io/friends/'
request=requests.get(primary_url)
soup=BeautifulSoup(request.content)
urllist=[]
for i in soup.find_all('a'):
    urllist.append(i.get('href'))
urllist=[primary_url+url for url in urllist]

```

```python
#appending the p in all the links to the final text list
text_list=[]
def processing_urls(urllist):
    for url in urllist:        
        newreq=requests.get(url)
        
        isoup=BeautifulSoup(newreq.content)
        text_list.append(isoup.find_all('p'))
processing_urls(urllist)
     
```

```python
import re
ross_list = []
rachel_list = []
chandler_list = []
joey_list = []
monica_list = []
phoebe_list = []

for i in text_list:
    for j in i:
        if re.search(r'Ross:', str(j)):
            ross_list.append(str(j))
        if re.search(r'Rachel:', str(j)):
            rachel_list.append(str(j))
        if re.search(r'Chandler:', str(j)):
            chandler_list.append(str(j))
        if re.search(r'Joey:', str(j)):
            joey_list.append(str(j))
        if re.search(r'Monica:', str(j)):
            monica_list.append(str(j))
        if re.search(r'Phoebe:', str(j)):
            phoebe_list.append(str(j))
chandler_list            
```

```python
clean_ross = []
clean_rachel = []
clean_chandler = []
clean_joey = []
clean_monica = []
clean_phoebe = []
def removetags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
i=0
for j in enumerate(chandler_list):
    text = removetags(chandler_list[i])
    clean_chandler.append(text)

    i+=1
i=0    
for j in enumerate(ross_list):
    text = removetags(ross_list[i])
    clean_ross.append(text)
    i+=1
i=0    
for j in enumerate(joey_list):
    text = removetags(joey_list[i])
    clean_joey.append(text)
    i+=1
i=0    
for j in enumerate(rachel_list):
    text = removetags(rachel_list[i])
    clean_rachel.append(text)
    i+=1
i=0    
for j in enumerate(phoebe_list):
    text = removetags(phoebe_list[i])
    clean_phoebe.append(text)
    i+=1
i=0    
for j in enumerate(monica_list):
    text = removetags(monica_list[i])
    clean_monica.append(text)
    i+=1      
    
```

```python
clean_list=[clean_ross,clean_monica,clean_chandler,clean_rachel,clean_phoebe,clean_joey]
for item in clean_list:
    i=0
    for line in item:
        line=line.replace('\n',' ')
        line=line.replace('\t',' ')
        item[i]=line
        i+=1
clean_monica        
```

```python
import io
num=0
with io.open('./Characters Transcript/ross list.txt','w',encoding='utf-8') as f:
    for ele in clean_ross:
        num+=1
        f.write(ele+'\n')
f.close()

num=0
with io.open('./Characters Transcript/Chandler list.txt','w',encoding='utf-8') as f:
    for ele in clean_chandler:
        num+=1
        f.write(ele+'\n')
f.close()

num=0
with io.open('./Characters Transcript/rachel list.txt','w',encoding='utf-8') as f:
    for ele in clean_rachel:
        num+=1
        f.write(ele+'\n')
f.close()

num=0
with io.open('./Characters Transcript/joey list.txt','w',encoding='utf-8') as f:
    for ele in clean_joey:
        num+=1
        f.write(ele+'\n')
f.close()

num=0
with io.open('./Characters Transcript/phoebe list.txt','w',encoding='utf-8') as f:
    for ele in clean_phoebe:
        num+=1
        f.write(ele+'\n')
f.close()

num=0
with io.open('./Characters Transcript/monica list.txt','w',encoding='utf-8') as f:
    for ele in clean_monica:
        num+=1
        f.write(ele+'\n')
f.close()
```
